# AED sur le dataset des fournisseurs

Dans cette section, nous allons effectuer une analyse exploratoire des données (AED) sur le dataset `fournisseurs`. Cette analyse nous permettra de mieux comprendre la structure des données, de détecter les valeurs manquantes, de visualiser les distributions des variables et d'identifier les relations potentielles entre les variables.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics

data = pd.read_csv('../RAW_data/fournisseursV2.csv', sep=";")
data.head()
```

```python
#on vient vérifier avec des boxplot si il y a des outliers (valeurs aberrantes)
plt.figure(figsize=(20,20))
plt.subplot(2,1,1)
sns.boxplot(x=data['fiabilité'])
plt.subplot(2,1,2)
sns.boxplot(x=data['délai_moyen_jours'])
plt.tight_layout()
plt.show()
```

Il n'y a pas d'outliers dans ces données

```python
numeric_cols = ['fiabilité', 'délai_moyen_jours']
corr = data[numeric_cols].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix')
plt.show()
```

On constate qu'il n'y pas de corélation entre les collones

# 🧴 Analyse Exploratoire des Produits

Ce notebook propose une **analyse descriptive** du fichier `produits.csv`, contenant les informations sur les produits en stock :
- ID produit
- Catégorie
- Marque
- Prix
- Stock minimum

---

## 🎯 Objectifs
- Comprendre la distribution des prix
- Identifier les marques et catégories dominantes
- Visualiser les seuils de stock minimum


```python
import pandas as pd

# Chargement
df = pd.read_csv("../RAW_data/produits.csv", sep=";", encoding="utf-8")
df.head()
```

## 📈 Statistiques générales

```python
df.describe(include="all")
```

## 💰 Distribution des prix des produits

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
plt.figure(figsize=(8, 4))
sns.histplot(df["prix"], bins=30, kde=True)
plt.title("Distribution des prix")
plt.xlabel("Prix (€)")
plt.ylabel("Nombre de produits")
plt.show()
```

## 🧴 Répartition par catégorie de produit

```python
plt.figure(figsize=(8, 4))
sns.countplot(data=df, y="catégorie", order=df["catégorie"].value_counts().index)
plt.title("Nombre de produits par catégorie")
plt.xlabel("Nombre de produits")
plt.ylabel("Catégorie")
plt.show()
```

## 🏷️ Marques les plus représentées

```python
top_marques = df["marque"].value_counts().head(10)
plt.figure(figsize=(8, 4))
sns.barplot(x=top_marques.values, y=top_marques.index)
plt.title("Top 10 des marques")
plt.xlabel("Nombre de produits")
plt.ylabel("Marque")
plt.show()
```

## 📦 Distribution du stock minimum

```python
plt.figure(figsize=(8, 4))
sns.histplot(df["stock_minimum"], bins=30, kde=True)
plt.title("Distribution du stock minimum")
plt.xlabel("Stock minimum")
plt.ylabel("Nombre de produits")
plt.show()
```

## ✅ Conclusion

Cette analyse nous apprend que :
- La majorité des produits ont un prix entre 250€ et 500€
- Certaines marques dominent largement l’assortiment
- La catégorie **Accessoires** semble très présente
- Les seuils de stock minimum varient fortement selon les produits

📌 Ces insights peuvent servir pour la **gestion des stocks**, le **référencement produit** ou la **détection des ruptures potentielles**.

## ðŸŽ¯ Objectifs
- Comprendre la distribution des prix
- Identifier les marques et catÃ©gories dominantes

```python
import pandas as pd

# Chargement
df = pd.read_csv("../RAW_data/produits.csv", sep=";", encoding="utf-8")
df.head()
```

## ðŸ“ˆ Statistiques gÃ©nÃ©rales

```python
df.describe(include="all")
```

## ðŸ’° Distribution des prix des produits

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
plt.figure(figsize=(8, 4))
sns.histplot(df["prix"], bins=30, kde=True)
plt.title("Distribution des prix")
plt.xlabel("Prix (â‚¬)")
plt.ylabel("Nombre de produits")
plt.show()
```

## ðŸ§´ RÃ©partition par catÃ©gorie de produit

```python
plt.figure(figsize=(8, 4))
sns.countplot(data=df, y="catÃ©gorie", order=df["catÃ©gorie"].value_counts().index)
plt.title("Nombre de produits par catÃ©gorie")
plt.xlabel("Nombre de produits")
plt.ylabel("CatÃ©gorie")
plt.show()
```

## ðŸ·ï¸ Marques les plus reprÃ©sentÃ©es

```python
top_marques = df["marque"].value_counts().head(10)
plt.figure(figsize=(8, 4))
sns.barplot(x=top_marques.values, y=top_marques.index)
plt.title("Top 10 des marques")
plt.xlabel("Nombre de produits")
plt.ylabel("Marque")
plt.show()
```

## ðŸ“¦ Distribution du stock minimum

```python
plt.figure(figsize=(8, 4))
sns.histplot(df["stock_minimum"], bins=30, kde=True)
plt.title("Distribution du stock minimum")
plt.xlabel("Stock minimum")
plt.ylabel("Nombre de produits")
plt.show()
```

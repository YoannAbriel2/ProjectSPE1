```python
import pandas as pd

achat = pd.read_csv('../RAW_data/achatsV2.csv', sep=";")
print(achat.head())
print("////////////////////////////////////////////////")


fournisseurs = pd.read_csv('../RAW_data/fournisseursV2.csv', sep=";") 
print(fournisseurs.head())
```

```python
df_merged = pd.merge(achat, fournisseurs, on="id_fournisseur", how="left")
df_merged.head()
```

```python
df_merged.to_csv('achats_merged.csv', index=False, sep=',')
```

```python
produits = pd.read_csv('../RAW_data/produitsV2.csv', sep=",")
stocks = pd.read_csv('../RAW_data/stocksV2.csv', sep=",")
df_produits_merge = pd.merge(produits, stocks, on="id_produit", how="left")
df_produits_merge.head()
df_produits_merge.to_csv('produits_merged.csv', index=False, sep=',')
```

```python
import pandas as pd

achat = pd.read_csv('../RAW_data/achatsV2.csv', sep=";")
print(achat.head())
print("////////////////////////////////////////////////")


fournisseurs = pd.read_csv('../RAW_data/fournisseursV2.csv', sep=";") 
print(fournisseurs.head())
```

```python
df_merged = pd.merge(achat, fournisseurs, on="id_fournisseur", how="left")
df_merged.head()
```

```python
df_merged.to_csv('achats_merged.csv', index=False, sep=',')
```

```python
produits = pd.read_csv('../RAW_data/produitsV2.csv', sep=";")
stocks = pd.read_csv('../RAW_data/stocksV2.csv', sep=";")

print(produits.head())
```

```python
df_produits_merge = pd.merge(produits, stocks, on="id_produit", how="left")

print(df_produits_merge.head())


df_produits_merge.to_csv('produits_merged.csv', index=False, sep=',')
```

```python
retour = pd.read_csv('../RAW_data/retoursV2.csv', sep=";")
commandes = pd.read_csv('../RAW_data/commandesV2.csv', sep=";")

commandes['id_commande'] = commandes['id_commande'].astype(str).str.strip().str.upper()
retour['id_commande'] = retour['id_commande'].astype(str).str.strip().str.upper()



retour = retour.drop(['id_produit'], axis=1)
retour.head()
```

```python
commandes.head()
```

```python
df_commandes_merged = pd.merge(commandes, retour, on="id_commande", how='left')

#remplissage des valeurs manquantes par null
df_commandes_merged = df_commandes_merged.fillna('null')

df_commandes_merged.to_csv('commandes_merged.csv', index=False, sep=',')
```
